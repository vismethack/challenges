

## Storm Eunice 2022
This notebooks should help participants to use the data from dataset for the storm Eunice from February 2022.

All the files are on ECMWF public server on this link: https://get.ecmwf.int/#browse/browse:vishackathon:eunice and can be browsed or downloaded.  
Grib data can be manually downloaded from here: https://get.ecmwf.int/service/rest/repository/browse/vishackathon/eunice/grib/  
And netcdf data from here: https://get.ecmwf.int/service/rest/repository/browse/vishackathon/eunice/netcdf/

Raw ECMWF data is in different units than some user might be used to. For example temperature is in Kelvins and mean sea level is in Pascals. Precipitation data is also stored as accumulated parameter.  
For this hackathon we have converted those into convenient units (C, hPa) and deacumulated precipitation.  
For the experienced users, we have also provided the file with the raw data in grib format.

## Download the data
Here's an example to download one file from grib directory using Python requests library.  
Other tools such as wget or curl can be used as well.

```python
import requests
filename = "eunice_10fg_fc_sfc.grib"
url = f"https://get.ecmwf.int/repository/vishackathon/eunice/grib/{filename}"
r = requests.get(url)
with open(filename, 'wb') as file:
    file.write(r.content)
```
